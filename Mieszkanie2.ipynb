```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

# Dane z treści zadania
initial_price = 120000  # początkowa cena mieszkania w złotych
annual_growth_rate = 0.05  # tempo wzrostu cen mieszkań
years = 5  # liczba lat

# Obliczanie przyszłej ceny mieszkania
future_price = initial_price * (1 + annual_growth_rate)**years

# Dane dotyczące lokaty
freq = 12
nominal_interest_rate = 0.12
monthly_periods = years * freq

# Obliczanie miesięcznej stopy procentowej
monthly_interest_rate = (1 + nominal_interest_rate)**(1/freq) - 1

# Obliczanie równomiernej miesięcznej wpłaty, aby zrównoważyć wzrost ceny mieszkania
monthly_payment = future_price / monthly_periods

# Generowanie danych dotyczących zmiany wartości lokaty i ceny mieszkania w czasie
months = np.arange(1, monthly_periods + 1)
price_increase = initial_price * (1 + annual_growth_rate / freq)**months

deposit_values = []
current_balance = 0

for month in range(1, monthly_periods + 1):
    current_balance += monthly_payment
    current_balance *= (1 + monthly_interest_rate)
    deposit_values.append(current_balance)

# Wypisanie wyników
print(f"Orientacyjna cena mieszkania za 5 lat: {future_price:.2f} zł")
print(f"Wymagana miesięczna wpłata do banku: {monthly_payment:.2f} zł")

# Generowanie wykresów
plt.figure(figsize=(10, 5))
plt.plot(months, price_increase, label='Cena mieszkania')
plt.plot(months, deposit_values, label='Wartość lokaty')
plt.xlabel('Liczba miesięcy')
plt.ylabel('Wartość (zł)')
plt.title('Zmiany w cenie mieszkania i wartości lokaty')
plt.legend()
plt.grid(True)
plt.show()
```
