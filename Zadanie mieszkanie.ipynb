```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

# Dane z treści zadania
initial_price = 120000  # początkowa cena mieszkania w złotych
annual_growth_rate = 0.05  # tempo wzrostu cen mieszkań
years = 5  # liczba lat

# Obliczanie przyszłej ceny mieszkania
future_price = initial_price * (1 + annual_growth_rate)**years

# Dane dotyczące lokaty
freq = 12
nominal_interest_rate = 0.12
monthly_interest_rate = nominal_interest_rate / freq
monthly_periods = years * freq

# Funkcja do obliczania miesięcznej wpłaty do banku
def calculate_monthly_deposit(present_value, future_value, periods, interest_rate):
    return npf.pmt(interest_rate, periods, present_value, future_value)

# Obliczanie wymaganej miesięcznej wpłaty
monthly_deposit = calculate_monthly_deposit(0, future_price, monthly_periods, monthly_interest_rate)

# Wypisanie wyników
print(f"Orientacyjna cena mieszkania za 5 lat: {future_price:.2f} zł")
print(f"Wymagana miesięczna wpłata do banku: {abs(monthly_deposit):.2f} zł")

# Generowanie wykresów
months = np.arange(1, monthly_periods + 1)
price_increase = initial_price * (1 + annual_growth_rate / freq)**months
deposit_values = np.full_like(months, abs(monthly_deposit))

plt.figure(figsize=(10, 5))
plt.plot(months, price_increase, label='Cena mieszkania')
plt.plot(months, deposit_values, label='Wpłata do banku')
plt.xlabel('Liczba miesięcy')
plt.ylabel('Wartość (zł)')
plt.title('Zmiany w cenie mieszkania i wpłatach do banku')
plt.legend()
plt.grid(True)
plt.show()
```
